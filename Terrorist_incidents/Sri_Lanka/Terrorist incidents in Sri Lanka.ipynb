```python
def srilanka():
    body = []
    for i in range(len(soup.select("table.wikitable.plainrowheaders.sortable.jquery-tablesorter"))):
        theads = soup.select("table.wikitable.plainrowheaders.sortable.jquery-tablesorter")[0].thead.text.split()
        table = soup.select("table.wikitable.plainrowheaders.sortable.jquery-tablesorter")[i]
        for r in table.find_all("tr"):
            body.append([i.text.strip() for i in r.find_all("td")])
    df = pd.DataFrame(body, columns = theads)
    df.drop([0,14,29,41], inplace=True)
    df.index = range(1,40)
    df["Date"] = pd.to_datetime(df["Date"])
    return df
```

```python
final_df = srilanka()
final_df
```
